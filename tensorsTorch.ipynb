### Math

```python
tensor1 = torch.tensor([1,2,3,4,5])
tensor2 = torch.tensor([6,7,8,9,10])
```

```python
# somma (somma di valori delle stesse posizione)
tensor1 + tensor2
torch.add(tensor1, tensor2)
```

```python
# sottrazione
tensor2 - tensor1
torch.sub(tensor2, tensor1)
```

```python
# prodotto
tensor1 * tensor2
torch.mul(tensor1, tensor2)
```

```python
# divisione
tensor2 / tensor1
torch.div(tensor2, tensor1)
```

```python
# modulo
tensor2 % tensor1
torch.remainder(tensor2, tensor1)
```

```python
# esponeziale
tensor1 ** tensor2
torch.pow(tensor1, tensor2)
```
